```python
def batch_weight_feature_and_nbatchpairs_scaling(strength, df_with_batch):
    nbs = df_with_batch.iloc[:, 0].nunique()
    w = nbs * (nbs - 1) / 2
    return(strength /( w * ( df_with_batch.shape[1] - 1 ) ) )
sys.path.append('../v1-DEBIAS-M-Analysis')
```

```python
res=pd.read_csv('hiv-age-results.csv', index_col=0).dropna() ## NAs are from the study w/ all positives
df_ = pd.read_csv('hiv-age-results-all.csv', index_col=0).dropna()
```

```python

df= pd.concat([res.loc[res.Group.isin(['Linear', 'DEBIAS-M'])], 
               df_])


df['Group']=df.Group.str.lower().str.replace('plsdabatch', 
                                             'plsda', 
                                             regex=False).str.replace('voom', 'snm')

order=[
       'conqur', 
       'percnorm',
       'snm',
       ]


data=df.copy()

data2=pd.read_csv('../results/HIVRC/auroc-boxplot-linear.csv', 
                 index_col=0
                 )
data['Covariate']='Age'
data2['Covariate']='Gender'

data_ = pd.concat([data[['auROC', 'Group', 'Covariate']], 
                   data2[['auROC', 'Group', 'Covariate']]
                  ]).reset_index(drop=True)


data_['Covariate_'] = data_['Group'] + ' ' + data_['Covariate']


global_palette['conqur Age'] = global_palette['conqur']
global_palette['percnorm Age'] = global_palette['percnorm']
global_palette['snm Age'] = global_palette['snm']
global_palette['conqur Gender'] = global_palette['conqur']
global_palette['percnorm Gender'] = global_palette['percnorm']
global_palette['snm Gender'] = global_palette['snm']

data_ = data_.loc[data_['Group'].isin(order)]
```

```python
plt.subplots(1,
                 figsize=(8, 10),
                 dpi=500)

ax=sns.boxplot(#x='Group', 
               y='auROC', 
               x='Covariate_',
               data=data_, 
               order=order_,
#                hue_order = order,
               palette=global_palette,
               width=.75,
               fliersize=0
           )

handles, labels = ax.get_legend_handles_labels()

sns.stripplot(#x='Group', 
              y='auROC', 
              x='Covariate_',
              data=data_,
              order=order_,
#               hue_order = order,
              dodge=True,
              ax=ax, 
              size=10, 
              color='k'
              )
# Put the legend out of the figure
 
# plt.legend(bbox_to_anchor=(1.01, .7), loc=2, borderaxespad=0.)
plt.legend().remove()
ax.set_title(None)
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.05))
plt.ylim(0.25, 1.05)
plt.xticks(rotation=90)
ax.set( yticks=[0.3,0.5,0.7, 0.9] )
plt.savefig('../results/covariate-selection/hiv-age-or-gender.pdf',
            dpi=900, 
            bbox_inches='tight',
            format='pdf')
plt.show()
```
