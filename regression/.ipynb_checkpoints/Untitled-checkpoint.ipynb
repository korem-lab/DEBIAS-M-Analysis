```python
base_pearsons=[]
dm_pearsons=[]
for i in range(cols.shape[0]):#cols.shape):
    
    if (md[cols[i]]>0 ).sum() > 200:
        np.random.seed(1)
        torch.manual_seed(1)
#         y = np.log10( md[cols[i]] + 1 ).values
        ss = StandardScaler()
        y=ss.fit_transform( np.log10( md[ [cols[i]] ] + 1 ) )[:, 0]
        X = rescale(df.values)

        
        ii = ( md[ cols[i] ]>0 ).values

        X=X[ii]
        grps=md.Study.values[ii]
        y=y[ii]

        inds = ( ( X > 1e-4 ).mean(axis=0) > .25 )#.25 )
#         inds = ( X ).max(axis=0) > .1# 1e-3

        X=rescale(X[:, inds])
        
        train_nm='HSRR_JR003_mm_corrected'
        train_inds = np.where( grps==train_nm )[0]
        test_inds = np.where( grps!=train_nm )[0]

#         logo=LeaveOneGroupOut()
#         for train_inds, test_inds in logo.split(X,
#                                                 y, 
#                                                 grps):
        lr=LinearRegression()
        lr.fit(X[train_inds], y[train_inds])
        print(y[train_inds].shape)

        
        base_pearsons.append( pearsonr(y[test_inds],
                             lr.predict(X[test_inds])
                             ) )

        X_with_batch=pd.DataFrame( 
                    np.hstack(( (grps==grps[0] ).astype(int)[:,np.newaxis], 
                                 X )) )


        dmr = DebiasMRegressor(x_val=pd.DataFrame( 
                                    X_with_batch.loc[test_inds] ), 
#                                mse_scaling=(1000.)/np.std(y), 
#                                learning_rate=1
                               )

        dmr.fit(pd.DataFrame( X_with_batch.loc[train_inds]), 
                y[train_inds])
        
        
        lr.fit( dmr.transform(X_with_batch).values[train_inds], 
               y[train_inds])


        dm_pearsons.append( pearsonr(lr.predict(dmr.transform(X_with_batch).values[test_inds]), 
                            y[test_inds]
                            ) )
```

```python
def batch_weight_feature_and_nbatchpairs_scaling(strength, 
                                                 df_with_batch):
    nbs = df_with_batch.iloc[:, 0].nunique()
    w = nbs * (nbs - 1) / 2
    return(strength /( w * ( df_with_batch.shape[1] - 1 ) ) )
```

```python
X=X[y>0]
grps=md.Study.values[y>0]
y=y[y>0]

logo=LeaveOneGroupOut()
for train_inds, test_inds in logo.split(X,
                                        y, 
                                        grps):
    lr=LinearRegression()
    lr.fit(X[train_inds][:, inds], y[train_inds])
    print(y[train_inds].shape)
    
    
    X_with_batch=pd.DataFrame( 
                np.hstack(( (grps==grps[0] ).astype(int)[:,np.newaxis], 
                     X[:, inds] )) )

    
    dmr = DebiasMRegressor(x_val=pd.DataFrame( 
                                X_with_batch.loc[test_inds]))
    
    dmr.fit(pd.DataFrame( X_with_batch.loc[train_inds]), 
            y[train_inds])
    break
```
