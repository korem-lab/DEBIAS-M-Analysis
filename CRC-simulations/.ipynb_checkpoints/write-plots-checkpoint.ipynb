```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import wilcoxon
import sys
sys.path.append('../v1-DEBIAS-M-Analysis/General_functions/')
import plotting
def flatten(xss):
    return [x for xs in xss for x in xs]
hue_ord=['Linear', 
         'combat', 
         'conqur', 
         'percnorm',
         'voom',
         'MMUPHin',
         'PLSDAbatch',
         'DEBIAS-M']
```

```python

res=pd.read_csv('res-crc-n-per-study.csv', index_col=0).fillna(0.5)

pall={'Linear':plotting.global_palette['linear'], 
                        'DEBIAS-M':plotting.global_palette['debias-m'],
                        'conqur':plotting.global_palette['conqur'],
                        'percnorm':plotting.global_palette['percnorm'],
                        'combat':plotting.global_palette['combat'], 
                        'MMUPHin':plotting.global_palette['mmuphin'], 
                        'PLSDAbatch':plotting.global_palette['plsda'], 
                        'voom':plotting.global_palette['snm']}

plt.figure(figsize=(12*1.2,12))
ax=sns.boxplot(y='auROC', 
               x='N per study',
               hue='Group',
               data=res,
               hue_order=hue_ord,
               fliersize=0,
               palette=pall
               
            )

sns.swarmplot(y='auROC', 
              x='N per study',
              hue='Group',
              color='black', 
              data=res,
              hue_order=hue_ord,
              s=5, 
              dodge=True,
              ax=ax
              )


sig_df = pd.DataFrame( flatten( [[ ( a,nps,
                           wilcoxon(res.loc[(res['N per study']==nps)&\
                                       (res['Group']=='DEBIAS-M')].auROC,
                               res.loc[(res['N per study']==nps)&\
                                       (res['Group']==a)].auROC, 
                               alternative='greater'
                              ).pvalue < 0.05
                    )
                        for a in res.Group.unique() if a != 'DEBIAS-M']
                   for nps in res['N per study'].unique()
                 ] ), 
             columns=['Group', 'N per study', 'sig_v_debias']
            )


x_val='N per study'
hue_val='Group'
sig_df['lower_y_val'] = min(res.auROC ) - 0.1
sns.swarmplot(x=x_val, 
              hue=hue_val,
              y='lower_y_val',
              data = sig_df.loc[sig_df.sig_v_debias], 
              hue_order = hue_ord,
              order=res['N per study'].unique(),
              marker='+',
              size=25/2, 
              ax=ax,
              palette=pall, 
              dodge=True, 
              color='black',
              linewidth=3.5/2
              )

sns.swarmplot(x=x_val, 
              hue=hue_val,
              y='lower_y_val',
              data = sig_df.loc[sig_df.sig_v_debias], 
              hue_order = hue_ord,
              order=res['N per study'].unique(),
              marker='x',
              size=17.5/2,
              ax=ax,
              palette=pall, 
              dodge=True, 
              color='black',
              linewidth=3.5/2,
              edgecolor="black"
              )

plt.ylabel('auROC') 
plt.yticks([.3, .5, .7, .9])
ax.legend().remove()
plt.savefig('../results/CRC-simulations/n-per-study.pdf', 
            format='pdf',
            dpi=900, 
            bbox_inches='tight'
            )

```
